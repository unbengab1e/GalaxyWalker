```python
from transformers import MllamaForCausalLM
import torch
```

```python
checkpoint = "/mnt/data/CVPR2025/task1_data/Llama-3.2-11B-Vision-Instruct"
model = MllamaForCausalLM.from_pretrained(checkpoint)
```

```python
cross_attention_layers = model.config.cross_attention_layers
```

```python
for name, param in model.named_parameters():
    for x in cross_attention_layers:
        if str(x) in name:
            param.requires_grad = True
            print(name)
        else:
            param.requires_grad = False
```
